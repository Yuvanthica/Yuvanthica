```python
df3 = pd.read_excel('/content/Resolution_TimeI.xlsx')
df4 = pd.read_excel('/content/Resolution_TimeII.xlsx')
```

```python
df5 = pd.merge(df3,df4)
```

```python
df5.head()
```

```python
df5['Volume'].isnull()
df5['Volume'].fillna(df5.Volume.mean(),inplace = True)

df5['Team_Experience'].isnull()
df5['Team_Experience'].fillna(df5.Team_Experience.mean(),inplace = True)

df5['Domain_Expertise'].isnull()
df5['Domain_Expertise'].fillna(df5.Domain_Expertise.mean(),inplace = True)

df5['Resolution_Time'].isnull()
df5['Resolution_Time'].fillna(df5.Resolution_Time.mean(),inplace = True)
```

```python
df5.isnull()
```

```python
df5.describe()
```

```python
df5.corr()
```

```python
sns.pairplot(df5,hue = 'Resolution_Time')
```

```python
from sklearn.preprocessing import StandardScaler as z
from sklearn.preprocessing import MinMaxScaler as minmax
```

```python
sns.pairplot(st_data,hue ='Resolution_Time')
```
