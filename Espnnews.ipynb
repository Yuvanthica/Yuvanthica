```python
# IMPORTS ----------------------------------------------------
import requests
from bs4 import BeautifulSoup
# ------------------------------------------------------------
keyword = input('Enter the word :')
# Requesting connection to website and collecting content ----
page_link = 'https://www.espn.in/search/_/q/'+ keyword
page_response = requests.get(page_link, timeout=5)
page_content = BeautifulSoup(page_response.content, "html.parser")
# ------------------------------------------------------------

counter = 0

for sectag in page_content.find_all('section', {'class': 'headlineStack__listContainer'}):
    for ultag in sectag.find_all('ul', {'class': 'headlineStack__list'}):
        if (counter < 10):
            for litag in ultag.find_all('li'):
                counter = counter + 1
                print(str(counter) + ". " + litag.text + " - http://www.espn.com" + litag.find('a')['href'])
        
```
