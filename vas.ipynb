```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro
from scipy.stats import stats
```

```python
df = pd.read_excel(r'/content/DATA+BHARATH.xlsx')
df.head()
```

```python
shapiro(df)
```

```python
df1 = df[['PRE A','POST A']]
df1
```

```python
shapiro(df1)
```

```python
df2 = df[['PRE B','POST B']]
df2
```

```python
shapiro(df2)
```

```python
df.describe()
```

paired t test between pre tests A nd B 

```python
stats.ttest_rel(df1['PRE A'],df2['PRE B'])
```

paired t test between post tests A nd B

```python
stats.ttest_rel(df1['POST A'],df2['POST B'])
```

independent t test within the group A

```python
stats.ttest_ind(df1['PRE A'],df1['POST A'])
```

independent t test between group  A nd B (pre test)

```python
stats.ttest_ind(df1['PRE A'],df2['PRE B'])
```

independent t test between group A nd B (post test)

```python
stats.ttest_ind(df1['POST A'],df2['POST B'])
```

independent t test within the group B

```python
stats.ttest_ind(df2['PRE B'],df2['POST B'])
```

paired t test between group B


```python
stats.ttest_rel(df2['PRE B'],df2['POST B'])
```

paired t test between group A


```python
stats.ttest_rel(df1['PRE A'],df1['POST A'])
```
